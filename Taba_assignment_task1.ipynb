```python
####Task 1 Data scrapping ###############################3

#Required Modules for Selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

#Required scraping and http requesting modules
from bs4 import BeautifulSoup
import requests

#Standard Library Modules for Data Manipulation
import time
import os
import csv
import sys
import pprint
from math import ceil
from operator import itemgetter

locationOfWebdriver = "C:\\Users\\YASH\\Desktop\\ISBASSIGN\\chromedriver"
driver = webdriver.Chrome(locationOfWebdriver)
driver2=webdriver.Chrome(locationOfWebdriver)

k=0
surl=str()
sfilename=str()
for k in range(0,3,1):
    if k == 0 :
        surl="https://www.cardekho.com/kia/seltos/user-reviews/"
        strmodel='KIA SELTOS'
        print("Scraping KIA SELTOS DATA")
        sfilename='C:\\Users\\YASH\\Desktop\\ISBASSIGN\\test\\cardekhokiascrapingtest.csv'
    if k ==1:
        surl="https://www.cardekho.com/mg/hector/user-reviews/"
        strmodel='MG HECTOR'
        print("Scraping MG HECTOR DATA")
        sfilename='C:\\Users\\YASH\\Desktop\\ISBASSIGN\\test\\cardekhohectorscrapingtest.csv'
    if k ==2:
        surl="https://www.cardekho.com/jeep/compass/user-reviews/"
        print("Scraping JEEP COMPASS DATA")
        strmodel='JEEP COMPASS'
        sfilename='C:\\Users\\YASH\\Desktop\\ISBASSIGN\\test\\cardekhocompassscrapingtest.csv'
    p=1
    with open(sfilename, 'w', encoding="UTF-8", newline="") as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(['Model','Title', 'Rating(scaled to 5)', 'Review', 'Reviewer','Date','url'])
    for p in range(1,7,1):
        if p == 1:
            url=surl
        else:
            url=surl+str(p)
        #print(url)
        driver.get(url)
        try:
            # find all the 10 jobs listed in the first page
            postlist = driver.find_elements_by_xpath("//div[@class='readReviewBox']/div[@class='readReviewHolder']")
            #print(postlist)
            lstrating=[]
            lstname=[]
            lstdate=[]
            lsttitle=[]
            lstreview=[]
            i=0
            for post in postlist:
                # get rating value
                try:
                    review = post.find_elements_by_xpath(".//div[@class='rating']/div[@class='starRating']/span/span")
                    t=0
                    #print(t)
                    for rating in review:
                        val = rating.get_attribute("class")
                        if(val.find('empty')==-1):
                            t+=1
                            #print(t)  
                except:
                    t='no title'
                lstrating.append(str(t))
                #get author name
                try:
                    authorname = post.find_element_by_xpath(".//div[@class=' authorInfo authordetail']/div[@class='authorSummary']/div[@class='name'] ")
                    an=authorname.text
                except:
                    an= 'ananymous user'
                lstname.append(an)
                #get posted date
                try:
                    postdate = post.find_element_by_xpath(".//div[@class=' authorInfo authordetail']/div[@class='authorSummary']/div[@class='date'] ")
                    pd=postdate.text
                except:
                    pd= 'dd-mm-yyyy'
                lstdate.append(pd)
                #get title
                try:
                    title = post.find_element_by_xpath(".//div[@class='contentspace']/h3")
                    head=title.text
                except :
                    head='no head'
                lsttitle.append(head)
                #getreveiw
                try:
                    temp=post.find_element_by_xpath(".//div[@class='contentspace']/h3/a")
                    lnk=temp.get_attribute('href')
                    driver2.get(lnk)
                    review=driver2.find_element_by_xpath("//div[@class='readReviewBox']/div[@class='readReviewHolder']/div[@class='contentspace']/p[@class='contentheight']")
                    re=review.text
                except :
                    re='reveiw   '
                lstreview.append(re)
                
            with open(sfilename,'a', encoding="UTF-8", newline="") as csv_file1:
                writer = csv.writer(csv_file1)
                for i in range(0,len(lstdate),1):
                    if p==2:
                        print(lsttitle[i])
                        print(lstrating[i])
                        print(lstname[i])
                    #print(i)
                    writer.writerow([strmodel,lsttitle[i],lstrating[i],lstreview[i],lstname[i],lstdate[i],surl[8:24]])
        except Exception as e:
            print(e)
            csv_file.close()
```
